## Методология

Вера в генетический и социальный детерминизм – адаптированные версии шкалы, которые необходимо конфиматорно подтвердить и оценить их психометрические свойства: надежность и валидности. 

**Конвергентная валидность:** 
- Шкала Правостороннего авторитаризма
- Шкала Социального доминирования 
- Шкалы имплицитных теорий личности и интеллекта 
- Шкала психологического эссенциализма
- Шкалы Большой Пятерки: Добросовестность и Доброжелательность

**Дифференцирующая:** 
- 2 утверждения из шкалы психологического эссенциализма

**Дискриминантная валидность:**
- Тест знаний в области генетики и социальных наук
- Субъективная оценка знаний (шкала уверенности) области генетики и социальных наук
- Опыт изучения генетики и социальных наук (вопрос из соц. дема)
- Шкалы Большой Пятерки: Экстраверсия, Эмоциональная стабильность и Открытость к Опыту

## Конфиматорный факторый анализ

### Основная логика
Будет построено 2 основных модели
1. **Общая модель:** без исключения утверждений и модификациий. Название = Модель 1

2. **Модель, сформированная исходя из результатов адаптации:**   
В опроснике некоторые утверждения были поделены на два:   
Пример общего утверждения: "Гены являются причиной, по которой родители и дети схожи в своем поведении и характере"    
   
   *Разделение:*   
   1) "Гены являются причиной, по которой родители и дети схожи в своем поведении"   
   2) "Гены являются причиной, по которой родители и дети схожи в своем характере"


Это нужно было, потому что респонденты не понимали, на что направлено утверждение. Следовательно, в первой модели оставляем все вопросы и самостоятельно добавляем модификации тех предложений, которые были разбиты на два. Название = Модель 2.1.

- Удаляем переменные с низкими факторными нагрузками + обратные вопросы (исходя из результатов корреляц. анализа). Название = Модель 2.2 
- Повторяем процедуру удаления в новой модели (при необходимости). Название = Модель 2.3 
- Еще раз повторяем процедуру удаления в новой модели (при необходимости). Название = Модель 2.4
- Добавляем модификации модели в модель 2.4. Модель 2.5.
- Сравниваем модели по фиту: 2.1; 2.2; 2.3; 2.4; 2.5


Последнее: нужно сравнить все получившиеся модели и выбрать лучшую, исходя из фитов   

Примечание: критерий низких факторных нагрузкок = < 0.04

## Модель №1: Общая

```r
#Дата-сет(новый)
factor <- gensoc

#Модель №1. Общая.
#Формула модели №1.
Model_base <- "
 Gen =~ BGD1 + BGD2 + BGD3 + BGD4 + BGD5 + BGD6 + BGD7 + BGD8 + BGD9 + BGD10 + 
 BGD11 + BGD12 + BGD13 + BGD14 + BGD15 + BGD16 + BGD17 + BGD18 + BGD19 + BGD20 +
 BGD21 + BGD22 + BGD23 + BGD24 
 
 Soc =~ BSD1 + BSD2 + BSD3 + BSD4 + BSD5 + BSD6 + BSD7 + BSD8 + BSD9 + BSD10 +
 BSD11 + BSD12 + BSD13"

# #Модель №1: использовать порядковые переменные + аргумент, который работает с ненормально распределением estimator = "WLSMV".
fit_Base <- cfa (Model_base, data = factor, missing = "listwise",estimator = 'WLSMV',
                  ordered=c("BGD1", "BGD2", "BGD3", "BGD4", "BGD5",
                            "BGD6", "BGD7", "BGD8", "BGD9", "BGD10",
                            "BGD11", "BGD12", "BGD13", "BGD14", "BGD15",
                            "BGD16", "BGD17", "BGD18", "BGD19", "BGD20",
                            "BGD21", "BGD22", "BGD23", "BGD24", "BSD1",
                            "BSD2", "BSD3", "BSD4", "BSD5", "BSD6",
                            "BSD7", "BSD8", "BSD9", "BSD10", "BSD11",
                            "BSD12", "BSD13"))
 
#Краткий аутпут 
fit_Base_scaled<- c("chisq.scaled", "df.scaled", 
                    "cfi.scaled", "tli.scaled", "rmsea.scaled",
                    "rmsea.ci.upper.scaled", "rmsea.ci.lower.scaled","srmr","aic","bic")  

round(fitMeasures(fit_Base, fit_Base_scaled),3)
```

**Результаты модели №1:**
- CFI - 0.703
- TLI - 0.685
- RMSEA - 0.108 (0.105-0.112)
- SRMR - 0.11

#### Факторные нагрузки первой модели

## Мультигрупповой конфиматорный факторный анализ 

#### Оценка инвариантности шкал по полу 

```r
#Добавляем пол в дата-сет по CFA
factor$Sex <- data$Sex

#Кол-во респондентов по полу
table(factor$Sex)
```

---
#### Конфигурационная инвариантность по полу

```r
fit1 <- cfa(Model_base10, factor, 
            estimator = 'MLR',
            group = 'Sex')

fit.scaled<-c("chisq.scaled", "df.scaled", 'p-value.scaled',
              "cfi.scaled", "tli.scaled", "rmsea.scaled",
              "rmsea.ci.upper.scaled", "rmsea.ci.lower.scaled","srmr","aic","bic")   

round(fitMeasures(fit1, fit.scaled),3)
```

#### Интерпретация 
1. Фиты очень близки к приемлемым. Самое главное, что RMSEA ок. 
2. Модель ок, поэтому можно переходить к метрической (слабой) инвариантности.
3. В целом, модель подходит для каждой из наших групп.

---
#### Метрическая (слабая) инвариантность по полу

```r
fit2 <- cfa(Model_base10, factor, 
            estimator = 'MLR',
            group = 'Sex',
            group.equal = "loadings")

fit.scaled<-c("chisq.scaled", "df.scaled", 
              "cfi.scaled", "tli.scaled", "rmsea.scaled",
              "rmsea.ci.upper.scaled", "rmsea.ci.lower.scaled","srmr","aic","bic")   

round(fitMeasures(fit2, fit.scaled),3)
```

#### Сравнение моделей: конфигурационная и слабая

```r
anova(fit2, fit1)
```

#### Фиты моделей: конфигурационная и слабая

```r
fit.scaled<-c("chisq.scaled", "df.scaled", 
              "cfi.scaled", "tli.scaled", "rmsea.scaled",
              "srmr","aic","bic")  

fit.stats <- rbind(fitmeasures(fit1, fit.measures = fit.scaled),
fitmeasures(fit2, fit.measures = fit.scaled))

rownames(fit.stats) <- c("configural", "weak invariance")
round(fit.stats,3)
```

#### Интерпретация 
1. У слабой модели фиты очень близки к приемлемым. Самое главное, что RMSEA ок. 
2. Мы не можем отклонить нулевую гипотезу. Это свидетельствует о слабой инвариантности.
3. Следовательно, можно сделать вывод, что факторы имеют одинаковое значение для каждой из групп (как свойство слабой инвариантности).

#### Скалярная (сильная) инвариантность по полу

```r
fit3 <- cfa(Model_base10, factor, 
            estimator = 'MLR',
            group = 'Sex',
            group.equal = c("loadings","intercepts"))

fit.scaled<-c("chisq.scaled", "df.scaled", 
              "cfi.scaled", "tli.scaled", "rmsea.scaled",
              "rmsea.ci.upper.scaled", "rmsea.ci.lower.scaled","srmr","aic","bic")   

round(fitMeasures(fit3, fit.scaled),3)
```

#### Сравнение моделей: сильная и слабая инвариантность
Мы сравниваем эти модели, поскольку модель сильной инвариантности вложена в слабую

```r
anova(fit3, fit2)
```

 P-value получился незначимым

#### Фиты моделей: конфигурационная, слабая и сильная

```r
fit.scaled<-c("chisq.scaled", "df.scaled", 
              "cfi.scaled", "tli.scaled", "rmsea.scaled",
              "srmr","aic","bic")  

fit.stats <- rbind(fitmeasures(fit1, fit.measures = fit.scaled),
fitmeasures(fit2, fit.measures = fit.scaled),
fitmeasures(fit3, fit.measures = fit.scaled))

rownames(fit.stats) <- c("configural", "weak invariance", "strong invariance")
round(fit.stats,3)
```

1. У модели сильной инвариантности фиты очень близки к приемлемым. Самое главное, что RMSEA ок.

3. Следовательно, можно сделать вывод, что мы имеем возможность сравнивать средние латентных переменных по полу.

#### Сравнение моделей: строгая и сильная инвариантность

```r
full <- parTable(fit5)#общее обозначение 
partial  <- subset(full, label == ".p38." | label == ".p50." |
                  label == ".p58." | label == ".p79.") #отберем вышеописанные пункты
partial
```

1. .p38.	==	.p130.  <-  это остаточная дисперсия (BGD3	~~	BGD3)

3. .p58.    ==  .p150.  <- это остаточная дисперсия (BSD9	~~	BSD9)

### Общая интерпретация процедуры мультигруппового конфиматорного факторного анализа 

Основной целью проведение процедуры мультигруппового конфиматорного факторного анализа - оценка инвариантности основных шкал по половой переменной. Это включало в себя проверку: конфигурационной, метрической, скалярной и строгой инвариантности.


Первоначально, была выполнена процедура оценки конфигурационной инвариантности. Параметры данной модели стремятся к конвенциональным значениям, χ = 932.82, df = 578, p <0.00; CFI = 0.89; TLI = 0.87; RMSEA = 0.055; SRMR = 0.069. Это говорит, что модель подходит для каждой из наших групп.


Далее, была проведена проверка метрической инвариантности. Параметры также стремятся к приемлемым показателям, χ = 947.49, df = 602, p <0.00; CFI = 0.89; TLI = 0.81; RMSEA = 0.053; SRMR = 0.072. 

Вдобавок, тест сравнения моделей (ANOVA) выявил статистически незначимый результат, χ_diff(602) = 16.05, p-value = 0.89. Следовательно, можно сделать вывод, что факторы имеют одинаковое значение для каждой из групп.

После этого была проведена оценка скалярной инвариантности. Параметры модели стремятся к приемлемым значениям, χ = 983.79, df = 626, p <0.00; CFI = 0.89; TLI = 0.88; RMSEA = 0.053; SRMR = 0.074.

Наконец, была выполнена оценка строгой инвариантности. Параметры этой модели стремятся к конвециональным значениям, χ = 1029.33, df = 652, p <0.00; CFI = 0.88; TLI = 0.88; RMSEA = 0.054; SRMR = 0.076.

Было обнаружено 3 параметра, которые необходимо было "освободить от равенства" в модели. Данные параметры отражены в следующих утверждениях: BGD3, BGD21, BSD9.

При повторном сравнении данной модели с моделью скалярной инвариантности была выявлена статистически незначимая разница, χ_diff(649) = 26.80, p-value = 0.026. Это говорит о возможности использовать суммарные баллы шкал для сравнения средних латентных переменных по полу, кроме утверждений BGD21, BGD3 и BSD9, поскольку вариации в ответах на данные утверждения среди мужчин и женщин различаются.
