## Методология

Вера в генетический и социальный детерминизм – адаптированные версии шкалы, которые необходимо конфиматорно подтвердить и оценить их психометрические свойства: надежность и валидности. 

**Конвергентная валидность:** 
- Шкала Правостороннего авторитаризма
- Шкала Социального доминирования 
- Шкалы имплицитных теорий личности и интеллекта 
- Шкала психологического эссенциализма
- Шкалы Большой Пятерки: Добросовестность и Доброжелательность

**Дифференцирующая:** 
- 2 утверждения из шкалы психологического эссенциализма

**Дискриминантная валидность:**
- Тест знаний в области генетики и социальных наук
- Субъективная оценка знаний (шкала уверенности) области генетики и социальных наук
- Опыт изучения генетики и социальных наук (вопрос из соц. дема)
- Шкалы Большой Пятерки: Экстраверсия, Эмоциональная стабильность и Открытость к Опыту 

**Дополнительно:**  шкала социальной желательности

## Основные вопросы

#### Описательная статистика 
1. Какие социально-демографические показатели у основной выборки?
2. Стремится ли выборка к репрезентативности?
3. Как распределились баллы у основных шкал?
4. Как связаны пункты между собой в основных шкалах?
5. Пригодны ли шкалы для факторного анализа?
6. Какие показатели надежности у основных шкал?

#### Конфиматорный факторный анализ
- Описываются ли основные шкалы двухфакторной моделью?

#### Мультигрупповой конфиматорный факторный анализ
- Обладают ли шкалы конфигурационной, слабой, сильной и строгой инвариативностью по переменной пол?

1. Гены являются причиной, по которой родители и дети схожи в своем поведении.
2. Гены являются причиной, по которой родители и дети схожи в своем характере.
3. Гены являются причиной алкоголизма.
4. Гены являются причиной, по которой мужчины и женщины различаются в своем поведении.
5. Гены являются причиной, по которой мужчины и женщины различаются в своем характере.
6. Гены являются причиной, по которой дети наследуют характер от родителей.
7. Гены являются причиной гомосексуальности.
8. Гены не влияют на поведение человека. (обратное)
9. Гены объясняют таланты человека.
10. Социальное окружение сильнее влияет на способности человека, чем его гены. (обратное)
11. Социальное окружение сильнее влияет на характер человека, чем его гены. (обратное)
12. Гены являются причиной, по которой люди с разным цветом кожи различаются в своем поведении.
13. Гены являются причиной, по которой люди с разным цветом кожи различаются в своем характере.
14. Гены не влияют на характер человека. (обратное)
15. Поведение человека – это врожденные инстинкты.
16. Гены определяют судьбу каждого человека.
17. Гены влияют на уровень интеллекта у человека.
18. Гены не объясняют развитие интеллекта у человека. (обратное)
19. Генетический анализ эмбриона (тест хромосом) позволяет предсказать характер ребенка в будущем.

23. Генетический анализ позволяет ученым предсказать способности человека, даже если они ничего не знают о нём.

### BSD - вера в социальный детереминизм:
1. Социальное происхождение человека можно определить по его характеру.
2. Прошлые социальные контакты человека (как длительные, так и краткосрочные) отражаются в его характере.

4. Социальное окружение сильнее влияет на поведение человека, чем его характер.
5. Социальное происхождение человека не влияет на его поведение. (обратное)

7. Социальное происхождение человека определяет его потенциальные возможности в жизни.
8. Социальное окружение человека влияет на его поведение.
9. Разные условия жизни являются причиной, по которой люди из богатых и бедных семей различаются по своему характеру.
10. Социальное происхождение человека не влияет на его характер. (обратное)
11. Социальное происхождение человека определяет его мысли.
12. Социальное происхождение человека определяет его поступки.
